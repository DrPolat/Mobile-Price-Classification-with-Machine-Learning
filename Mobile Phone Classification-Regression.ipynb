## <font color="red"> Regression

### preparing the data

```python
train=df[:2000]
```

```python
train.tail()
```

```python
test=df[2000:]
```

```python
test.head()
```

```python
x=train.drop("price_range",axis=1)
y=train["price_range"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import  ExtraTreeRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.linear_model import Ridge, Lasso, ElasticNet #ikinci bir kaynaktan başarının doğruluğunu ölçmeye yarıyor
```

```python
def algo_test (x,y):
    l=LinearRegression()
    e=ElasticNet()
    r=Ridge()
    lass=Lasso()
    rfr = RandomForestRegressor()
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20, random_state=42)
    
    algos= [l,e,r,lass,rfr]
    algo_names = ["Linear","ElasticNet","Ridge","Lasso","RandomForestRegressor"]
    r_squared= []
    rmse = []
    mae= []
    
    result = pd.DataFrame(columns=["R_Squared","RMSE","MAE"],index = algo_names)
    
    for algo in algos:
        algo.fit(x_train,y_train)
        r_squared.append(r2_score(y_test,algo.predict(x_test)))
        rmse.append(mean_squared_error(y_test, algo.predict(x_test))**.5)
        mae.append(mean_absolute_error(y_test, algo.predict(x_test)))
    
     
        
    result.R_Squared = r_squared
    result.RMSE = rmse
    result.MAE = mae
    
    return result.sort_values("R_Squared", ascending=False)
```

```python
algo_test(x,y)
```
